Encima disto poderiamos fazer um calculo simples, como por exemplo

```r
a=b=1
print(a)
print(b)
for(i in 3:10){
    c=a+b
    a=b
    b=c
    print(c)
}
```

Para saber os n elementos desta sequencia podemos querer saber quais são os fatoriais de n elementos como

```r
for(i in 0:10){
    print(factorial(i))
    }
```

tambem podemos utilizar a estrutura condicional while

```r
c=1
cont=0
while(c<1e5){
    c=factorial(cont)
    print(c)

    cont=cont+1
}
```

Claramente vemos que o while avalia uma expressão logica, assim poderiamos ter passado variaveis logicas, tambem é valido Construir strings utilizando um loop, por exemplo

```r
StringIni="Letras:"
for(i in 1:26){
    StringIni=paste(StringIni,letters[i],sep=",")
    print(StringIni)
}
```
