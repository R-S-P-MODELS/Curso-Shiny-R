Neste notebook discutiremos como trabalhar com matrizes na linguagem R, matrizes aqui são representadas apenas com valores numericos, para trabalhar com valores categoricos será nescessario o dataframe

As operações e representação de matrizes em R vem da representação em fortran, e o codigo executado por tras das funções é primeiramente compilado do fortran

```r
Matriz=matrix(runif(25),nrow=5,ncol=5)
Matriz
```

```r
dim(Matriz)
```

```r
Matriz2=matrix(rnorm(25),nrow=5,ncol=5)
Matriz2
```

Como previamente dito, as operações em R são executaveis se podem ser corretamente definidas desde que definida para a estrutura de dados, assim

```r
#Soma
Matriz+Matriz2
```

```r
#Subtração
Matriz-Matriz2
```

```r
#Produto Tensorial
Matriz*Matriz2
```

```r
#Produto de matrizes
Matriz %*% Matriz2
```

```r
#Determinante do produto
det(Matriz %*% Matriz2)
```

```r
#Transposta da matriz
t(Matriz)
```

```r
# Alem disto podemos calcular a matriz inversa ou resolver um sistema linear Ax=b
```

```r
Inversa=solve(Matriz %*% Matriz2)
Inversa
```

```r
Identidade=Inversa %*% Matriz %*% Matriz2
Identidade
```

Vemos que esta é a matriz identidade, podemos gerar um vetor B então para calcular Ax=B

```r
B=rexp(5,3)
B
```

```r
x=solve(Matriz %*% Matriz2,B)
```

```r
Erro=Matriz %*% Matriz2 %*% x -B
```

```r
Erro
```

```r
mean(Erro^2)
```

Observamos que o metodo obtem otimos resultados mesmo com valores aleatorios, mostramos que ele é escalavel utilizando uma matriz 1000 x 1000 aleatoria

```r
A=matrix(rnorm(1e6,5,3),1e3,1e3)
B=runif(1e3,-5,5)
x=solve(A,B)
```

```r
Erro=A %*%x -B
```

```r
mean(Erro^2)
```

Mesmo com uma matriz 1000x1000 gerada aleatoriamente, vemos que a solução alem de rapida tem otimos resultados

Podemos testar o gerador calculando a correlação de pearson desta matriz de numeros aleatorios por

```r
correlação=cor(A)
mean(correlação^2)
```

Onde valores proximos de 0 significa que o gerador é bom ja que as variaveis aleatorias devem ser descorrelacionadas

Alem disto matrizes podem aplicar algumas funções sobre linhas ou colunas, por exemplo

```r
colMeans(A)
```

```r
rowMeans(A)
```

que representa a media calculada por coluna e por linha da matriz A
