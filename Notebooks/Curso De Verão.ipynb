Este notebook tem como proposito apresentar informações sobre a linguagem R, e alguns exemplos de como utilizala

Primeiramente apresentamos os tipos de variaveis contidos na linguagem por padrão:

Estas são variaveis inteiras, numericas(reais) caracteres, fatores, e logicas

R é uma linguagem em que não é nescessario a declaração do tipo de variavel, e que uma mesma variavel pode ter seu tipo mudado continuamente durante o programa, como por exemplo

```r
Inteira<-as.integer(3)
print(Inteira)
print(class(Inteira))

Real<- 3.0
print(Real)
print(class(Real))

Caracter <- '3'
print(Caracter)
print(class(Caracter))

Fator=as.factor(3)
print(Fator)
print(class(Fator))

Logico=3<2
print(Logico)
print(class(Logico))
```

Existem operações especificas associadas a cada tipo, alem de restrições associadas, as variaveis inteiras e numericas por exemplo permitem operações matemáticas sobre estes, caracteres são utilizados para compor palavras e a diferença entre este e fatores e que dado um vetor de fatores, não é possivel inserir um elemento neste vetor que já não esteja definido neste, isto permite controle quando é do desejo do usuário representar um atributo cujo contem um conjunto finito especifico de valores aceitaveis a serem representados. As variaveis logicas são utilizadas para as operações logicas como AND, OR, NOT, entre outras, alguns exemplos de operações entre estas seguem a seguir

```r
Inteira +Real
paste(Caracter,Caracter,sep="")
length(levels(Fator))
Logico | !Logico
```

```r
Inteira=Real
print(Inteira)
```

Tendo então variaveis declaraveis, podemos apresentar as estruturas de Loop

Apresentamos a seguir a estrutura de Loop For, e a estrutura while, que funcionam de forma similar a outras linguagens

```r

for(i in seq(0, 10, by = 2)){
    print(2*i)
}
```

```r
a=1
b=1
c=a+b
while(c<100){
    print(c)
    a=b
    b=c
    c=a+b
    
}
```

A estrutura de loop se mantem desta forma, mas uma propriedade da linguagem R a ser demonstrada a seguir e que este processo de loop deve ser evitado, uma vez que R é uma linguagem que aplica suas funções em vetores e matrizes naturalmente. Primeiramente apresentamos a criação de um vetor na linguagem R

```r
# Criando um vetor vazio
Vetor=vector()
print(Vetor)
Vetor[1]=1
Vetor[2]=2
print(Vetor)

# Criando um vetor da sequencia 1:100
Vetor=1:100

print(Vetor)
```

Uma das propriedades da linguagem R e que não precisamos de estruturas de loop para fazer operações sobre estes vetores, como vemos a seguir

```r
2*Vetor
sum(Vetor)
min(Vetor)
Vetor/2
summary(Vetor)
```

Alem disto podemos gerar vetores a partir de valores aleátorios gerados por diferentes distribuições de probabilidade, e gerar gráficos de suas distribuições

```r
#Distribuição Uniforme
Vetor<-runif(n = 1000,min = 0,max = 1)
hist(Vetor,main = "Distribuição Uniforme")

#Distribuição Gaussiana
Vetor<-rnorm(n = 1000,mean = 0,sd = 1)
hist(Vetor,main = "Distribuição Normal")

#Distribuição Exponencial
Vetor<-rexp(n = 1000,rate = 0.5)
hist(Vetor,main = "Distribuição Exponencial")
#Distribuição Chi Quadrada
Vetor<-rchisq(n = 1000,df = 2,ncp = 1)
hist(Vetor,main = "Distribuição Chi Quadrada")
```

Podemos verificar como o acumulado de vetor se comporta tambem utilizando-se da função cumsum

```r
plot(cumsum(Vetor),xlab = "Indice",ylab="Somatorio",main = "Vizualização")
```

Aqui falaremos um pouco sobre funções
Funcoes em R não requerem que as variaveis tenham um tipo declarado, o que implica que se as operações contidas na função forem definidas para mais de um tipo de variavel estas podem ser utilizadas

Por exemplo vamos fazer uma funcao que retorna a multiplicado por b a seguir

```r
Produto<-function(a,b){
    return(a*b)
}
```

```r
a=1
b=2
Produto(a,b)
```

```r
a=1:10
b=2*a
Produto(a,b)
```

```r
a=matrix(1:9,3,3)
b=a/3
Produto(a,b)
```

```r
Produto(2,b)
```

```r
a<-'a'
b='b'
Produto(a,b)
```

Como vemos as funções não tem tipo fixo nem para receber ou para retornar, desde que as operações contidas nela façam sentido para o tipo de variavel especificada

Para funcoes retornarem varias informações, esta deve ser retornada em uma lista, façamos uma função que retorna Media Mediana Variancia minimo e maximo

```r
Estatisticas<-function(Variavel){
    Est<-list()
    Est[[1]]=mean(Variavel)
    Est[[2]]=median(Variavel)
    Est[[3]]=var(Variavel)
    Est[[4]]=min(Variavel)
    Est[[5]]=max(Variavel)
    names(Est)=c('Media','Mediana','Variancia','Minimo','Maximo')
    return(Est)
}
```

```r
vec<-rnorm(1e4,mean = 5,sd = 10)
Dados<-Estatisticas(vec)
```

```r
Dados$Media
Dados$Mediana
Dados$Variancia
Dados$Minimo
Dados$Maximo
```

```r
hist(vec)
```

Normalmente esta função poderia operar sobre matrizes tambem

```r
Mat<-matrix(1:1e4,1e2,1e2)
MatEst<-Estatisticas(Mat)
MatEst
```

Em seguida teremos a demonstração das funções sapply e lapply para dados do Kickstarter

```r
lapply(lapply(Kickstarter,unique),length)
```

Os codigos acima mostram respectivamente a quantidade de valores unicos para cada coluna, e sua proporção em relação ao numero total de valores do dataset para cada atributo,a seguir para o vapply

```r
u=which(sapply(Kickstarter,class) %in% c('integer','numeric'))
names(Kickstarter)[u] # variaveis numericas do kickstarter
KickNumerico=Kickstarter[u]
vapply(X = KickNumerico,mean,FUN.VALUE = 0)

```

A funcao vapply se torna uma otimização da função sapply quando todas as saidas retornam variaveis de mesma classe, neste caso é requerido um exemplo de uma variavel deste tipo no atributo FUN.VALUE

https://pt.stackoverflow.com/questions/8056/apply-sapply-mapply-lapply-vapply-rapply-tapply-replicate-aggregate-by-e

```r
#A função mapply é utilizada sobre listas que contem estruturas de dados diferentes
names(NovaLista)
mapply(NovaLista,FUN = length)
```

```r
#rapply aplica uma funcao recursiva sobre cada elemento de uma lista
rapply(NovaLista,f=length)
```

E quando queremos juntar todos os elementos de uma lista em um unico elemento, caso uma operação seja valida para todos

Neste caso podemos utilizar as funções do.call ou Reduce que operam sobre uma lista de n elementos retornando apenas um caso seja possivel a seguir vemos alguns exemplos

```r
Lista=list()
Lista[[1]]=1:10
Lista[[2]]=matrix(1:9,3,3)
Reduce(f = sum,Lista)
do.call(what = sum,Lista)
```

```r
Lista[[1]]=Kickstarter
Lista[[2]]=Kickstarter
Kick2=Reduce(rbind,Lista)
dim(Kickstarter)
dim(Kick2)
```
