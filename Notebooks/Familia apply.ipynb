Até então  sempre que não pudemos aproveitar as propriedades vetoriais do R, estivemos utilizando loops

A questão e que utilizar se de estruturas de loop não é escalavel na linguagem fazendo então com que o codigo se torne lento quando o tamanho de uma amostra aumenta como exemplo para demonstrar isto utilizemos o comando System

```r
med=c()
i=1e6
m=matrix(runif(3*i),nrow=3)
    for(j in 1:ncol(m)){
        med[j]=mean(m[,j])
    }
  
```

Uma segunda forma de fazer este loop interno de j e utilizando a função apply

apply é uma função usavel em matrizes ou estruturas similares

a sintaxe dela é a seguinte:

apply(Dado,referencia,função)

onde dado é sua matriz, referencia é o valor 1 ou 2 que indica se a funcao será aplicada sobre as linhas ou colunas e funcao é a funcao em questão utilizada, refazendo o bloco anterior com apply

```r
i=1e6
m=matrix(runif(3*i),nrow=3)
med=apply(m,2,mean)
```

Caso quisessemos aplicar a função iterando sobre as linhas deveria se executar

```r
med=apply(m,1,mean)
```

Um outro tipo de função aplicada quando não estamos pensando em uma estrutura de matriz são as seguintes:

lapply (List apply)
sapply (Simple apply)
vapply
mapply (Multiple apply)

Para trabalhar com estas instanciemos uma lista que contenha um matriz um vetor e um dataframe

```r
Lista=list(vec=rexp(100,3),matriz=matrix(runif(100),10,10),Frame=read.csv("Kickstarter.csv"))
```

```r
names(Lista)
```

Nesta lista poderiamos aplicar varios exemplos de funções apply, por exemplo

```r
apply(Lista$matriz,2,mean)
```

Nos exemplos anteriores utilizamos loops para verificar a classe de cada elemento do dataframe, podemos utilizar a função lapply ou sapply para realizar este processo

```r
sapply(Lista$Frame,class)
```

```r
lapply(Lista$Frame,class)
```

Vemos a diferença entre lapply e sapply e que lapply sempre retorna uma lista, enquanto  sapply tenta retornar a saida em uma forma mais simples

Poderiamos combinar estas e ao inves de utilizar um loop duplo, utilizar dois sapply um dentro do outro

```r
sapply(sapply(Lista$Frame,unique),length)
```

Observamos o numero de elementos unicos para cada coluna, sem a nescessidade de implementar 2 loops duplos

Caso a saida do processo analitico seja sempre do mesmo tipo, o usuario pode otimizar a analise utilizando a função vapply, esta funciona como a função lapply, mas requer como argumento um elemento de mesma classe da saida (como por exemplo 1 para saida numerica , "a" para caracteres) etc

```r
vapply(X = Lista$Frame,FUN = class,FUN.VALUE = "a")
```

```r
vapply(Lista$matriz,mean,FUN.VALUE = 1)
```

E tambem introduzimos a função mapply( Esta deve operar sobre listas que não possuem o mesmo tipo de elemento em seus indices

```r
mapply(Lista,FUN=length)
```

A função torna como referencia que length calcula tamanho para uma linha fixa, então retornou o numero de colunas de matriz e Frame

Tambem possuimos a função tapply, esta tem como propriedade que o calculo é feito agrupado por um identificador do tipo fator

è comumente usado quando queremos fazer um calculo por grupos

```r
tapply(Lista$Frame$backers_count,Lista$Frame$country,mean)
```

Este é o numero medio de pessoas que apoiam um projeto por pais
