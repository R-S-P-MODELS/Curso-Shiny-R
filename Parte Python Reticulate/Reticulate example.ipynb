```r
library(reticulate)
pandas = import("pandas",convert=FALSE)
titanic = pandas$read_csv("https://goo.gl/4Gqsnz")
titanic$describe()
```

```r
caracteres=which(sapply(py_to_r(titanic),class) %in% 'character')
titanic_r=py_to_r(titanic)
```

```r
for(i in caracteres)
    titanic_r[,i]=as.numeric(as.factor(titanic_r[,i]))
```

```r
#titanic_r=py_to_r(titanic)
titanic_r=titanic_r[,2:10]
titanic_r=titanic_r[,-3]
```
